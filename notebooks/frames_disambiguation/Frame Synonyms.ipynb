```python
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, join

frames_folder = "/home/anca/Documents/frames/data/crowdsourcing/output/syn-frames/amt_strict/"
synfiles = [f for f in listdir(frames_folder) if isfile(join(frames_folder, f))]

all_syns = set()

for f in range(len(synfiles)):
    input_syns = pd.read_csv(frames_folder + synfiles[f])
    syn_list = [str(s).split("|") for s in input_syns["Answer.SynPhrase"]]
    flatten = np.unique([item.lower().replace(" ", "_") for sublist in syn_list for item in sublist])
    all_syns = all_syns.union(flatten)

all_syns = list(all_syns)
all_syns = [str(s).replace('none_of_the_above.', 'none') for s in all_syns]
# 'none' in all_syns
```

```python
#frames_folder = "/home/anca/Documents/frames/data/crowdsourcing/output/syn-frames/amt_strict/"

#frames_dataset.head(n=10)
import sys
stdout = sys.stdout

sys.path.append('../../')

from defaultconfig import Configuration

class FramesConfig(Configuration):
    inputColumns = ["Input.word_phrase", "Input.lemmatized", "Input.pos", "Input.beg", "Input.end", "Input.nr_syns",
                    "Input.frame_syns", "Input.sentence", "Input.syn_sents", "Input.sent_id"]
    outputColumns = ["Answer.SynPhrase"]
    
    # processing of a closed task
    open_ended_task = True
    annotation_vector = all_syns
    
    def processJudgments(self, judgments):
        # change default separator to whitespace to make it work with our file
        for col in self.outputColumns:
            judgments[col] = judgments[col].apply(lambda x: str(x).replace('|',','))
            
            judgments[col] = judgments[col].apply(lambda x: str(x).replace('None of the above.','none'))
            judgments[col] = judgments[col].apply(lambda x: str(x).replace(' ','_'))
            judgments[col] = judgments[col].apply(lambda x: str(x).lower())
        return judgments

config = FramesConfig()

# pre-process the data and create the annotation vectors
from controllers.inputController import processFile
pre_processed_results = processFile(
    root=".", directory=frames_folder, filename="",
    config=config
)
```

```python
# %%debug

sys.stdout = stdout

# run the metrics
from models import Metrics
processed_results = Metrics.run(pre_processed_results, config)
```

```python
unique_sentences = processed_results["units"]["input.sentence"]
unique_syns = config.annotation_vector

sentences = []
syns = []
words = []
sent_text = []
sss = []
sent_ids = []


for idx in processed_results["units"].index:
    sent_syns = processed_results["units"]["input.frame_syns"][idx].split(",")
    sent_syns = [x.replace(' ','_') for x in sent_syns]
    for syn in sent_syns:
        sentences.append(idx)
        sent_ids.append(processed_results["units"]["input.sent_id"][idx])
        words.append(processed_results["units"]["input.word_phrase"][idx])
        sent_text.append(processed_results["units"]["input.sentence"][idx])
        syns.append(syn)
        if syn in processed_results["units"]["unit_annotation_score"][idx].keys(): 
            sss.append(processed_results["units"]["unit_annotation_score"][idx][syn])
        else:
            sss.append(0.0)
syn_sentence_score = pd.DataFrame({
    "sent_id" : sent_ids,
    "sentence" : sent_text,
    "word" : words,
    "syn_sent_score" : sss,
    "syn_word" : syns
})

syn_sentence_score
```
