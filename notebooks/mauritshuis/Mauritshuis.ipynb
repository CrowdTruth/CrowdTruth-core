```python
import pandas as pd

ques_paint_filename = "workers-vector.csv"
ques_paint_dataset = pd.read_csv(ques_paint_filename, sep = ";")

ques_paint_dataset
```

```python
import sys
import numpy as np

stdout = sys.stdout

sys.path.append('../../')

from defaultconfig import Configuration

class QuesPaintConfig(Configuration):
    inputColumns = ["question_id", "question_text", "total_number_of_boxes"]
    outputColumns = ["judgments"]
    
    # processing of an open-ended task
    open_ended_task = True
    column_separator = ';'
    
    def processJudgments(self, judgments):
        dlist = []
        for idx in judgments.index:
            judg_dict = {}
            for jdx in range(judgments["total_number_of_boxes"][idx]):
                judg_dict["score-region/" + str(jdx)] = judgments["score-region/" + str(jdx)][idx]
            judg_dict["nothing-to-box"] = float(judgments["nothing-to-box"][idx])
            dlist.append(judg_dict)
        judgments["judgments"] = dlist
        judgments = judgments.replace(np.nan, '', regex=True)
        return judgments

config = QuesPaintConfig()

# pre-process the data and create the annotation vectors
from controllers.inputController import processFile
pre_processed_results = processFile(
    root=".", directory="", filename=ques_paint_filename,
    config=config
)
```

```python
sys.stdout = stdout

# run the metrics
from models import Metrics
processed_results = Metrics.run(pre_processed_results, config)
```
